## Answeting question 3

- We use subset to remove species from the data table as it is not a variable we can work with

```r
mydata = subset(iris, select = -c(Species))
```

```r
df_1 = mydata %>% 
  mutate(Sepal.Length = replace(Sepal.Length, 1, NA))
head(df_1)
```

```r
df_2 = mydata %>% 
  mutate(
    Sepal.Length = replace(Sepal.Length, 1, NA),
    Sepal.Width = replace(Sepal.Width, 2, NA),
    Petal.Length = replace(Petal.Length, 3, NA),
    Petal.Width = replace(Petal.Width, 4, NA),
)
 
head(df_2) 
```

---
- Missing values are a problem in predictive modelling and genrally they can undermine the accuracy of our dataset with imputation we can generate values for our missing values in this task we use "knnImpute" and "bagImpute".

- preProcess can be used to impute data sets based only on information in the training set. One method of doing this is with K-nearest neighbors. For an arbitrary sample, the K closest neighbors are found in the training set and the value for the predictor is imputed using these values (e.g. using the mean). Using this approach will automatically trigger preProcess to center and scale the data, regardless of what is in the method argument.
- We use "knnImpute" identify and replace missing values for each column in our input data prior to modeling for prediction task.
- It replaces data with nearest neighbor estimated values

```r
model <- "knnImpute"

ppparams <- preProcess(df_1, method = model)
train <- predict(ppparams, mydata)
summary(train)
```

```r
ppparams <- preProcess(df_2, method = model)
train <- predict(ppparams, mydata)
summary(train)
```

### OUTPUT

- A new sample is imputed by finding the samples in the training set “closest” to it and averages these nearby points to fill in the value
- From the data observed it is viewed that the missing value was estimated from the meand of the values surrounding it 
- we can suggest that "knnImpute" averages (mean) value of closest values.

----
- Alternatively, bagged trees can also be used to impute. For each predictor in the data, a bagged tree is created using all of the other predictors in the training set. When a new sample has a missing predictor value, the bagged model is used to predict the value. While, in theory, this is a more powerful method of imputing, the computational costs are much higher than the nearest neighbor technique.



```r
model2 <- "bagImpute"

ppparams <- preProcess(df_1, method = model2)
train <- predict(ppparams, mydata)
summary(train)
```

```r

ppparams <- preProcess(df_2, method = model2)
train <- predict(ppparams,mydata)
summary(train)
```

- Scaling may be important when you’re trying to sum or average variables that are on different scales, perhaps to create a composite score of some kind. Without scaling, it may be the case that one variable has a larger impact on the sum due purely to its scale, which may be undesirable.

```r
processParams <- preProcess(irist, method = c("scale"))
print(processParams)

training = predict(processParams, irist)

ggplot(data=training, aes(Sepal.Length)) +
    geom_histogram(bins = 15)
qqnorm(training$Sepal.Length, pch=1, frame=FALSE, main='scaled qqplot of sepal length')
qqline(training$Sepal.Length, col='blue', lwd = 1)
```

- The plot doesnt look very much like a noraml plot in the histogram but judging from the quantile plot we can see that the plot is normal is a bit of skewness.

---

- Centering may be needed to simplify calculations and notation. For example, the sample covariance matrix of a matrix of values centered by their sample means is simply X′X. Similarly, if a univariate random variable X has been mean centered, then var(X) = E (X<sub>2</sub>) and the variance can be estimated from a sample by looking at the sample mean of the squares of the observed values.

```r
processParams <- preProcess(irist, method = c("center"))
print(processParams)

training = predict(processParams, irist)

ggplot(data=training, aes(Sepal.Length)) +
    geom_histogram(bins = 15)
qqnorm(training$Sepal.Length, pch=1, frame=FALSE, main='scaled qqplot of sepal length')
qqline(training$Sepal.Length, col='blue', lwd = 1)
```

- The plot doesnt appear to be normal from the histogram but judging from the quantile plot we can see that the plot is normal is a bit of skewness.
- Different histogram plot indicating a drop towards zero in the centre of the plot which could be suggesting a zero mean value.
- Similar to the first plot without any transformations in the qq plot.
- The sample quantiles on the y axis is reduced from 8 to -1.5 to 2 due to zero mean which indicates a normalized data. 

```r
processParams <- preProcess(irist, method = c("scale","center"))
print(processParams)

training = predict(processParams, irist)

ggplot(data=training, aes(Sepal.Length)) +
    geom_histogram(bins = 15)
qqnorm(training$Sepal.Length, pch=1, frame=FALSE, main='scaled qqplot of sepal length')
qqline(training$Sepal.Length, col='blue', lwd = 1)
```

- The plot is more similar to the center plot showing a zero mean behaivour this is highly possible beacuse the mean is divided by the standard deviation.
- From the qq plot we can suggest it is from a normal distribution as it falls on the line with some skewness.
